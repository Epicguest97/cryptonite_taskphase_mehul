```python
import numpy as np
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA  # For initial dimensionality reduction

class FastTSNE:
    def __init__(self, n_components=2, perplexity=30.0, learning_rate=200.0, 
                 n_iter=300, random_state=None):
        self.n_components = n_components
        self.perplexity = perplexity
        self.learning_rate = learning_rate
        self.n_iter = n_iter
        self.random_state = random_state
        
    def fit_transform(self, X):
        if self.random_state is not None:
            np.random.seed(self.random_state)
            
        # Initial dimensionality reduction using PCA to 50 dimensions
        if X.shape[1] > 50:
            pca = PCA(n_components=50)
            X = pca.fit_transform(X)
            
        n_samples = X.shape[0]
        
        # Compute squared distances (vectorized)
        sum_X = np.sum(X**2, axis=1)
        D = -2 * np.dot(X, X.T) + sum_X.reshape((-1, 1)) + sum_X
        
        # Compute P matrix (high-dimensional affinities)
        P = self._compute_p_matrix_fast(D)
        
        # Initialize low-dimensional embeddings
        Y = np.random.normal(0, 1e-4, (n_samples, self.n_components))
        
        # Initialize momentum parameters
        Y_prev = Y.copy()
        gains = np.ones_like(Y)
        
        # Early exaggeration
        P = P * 4.0
        
        # Gradient descent with momentum and adaptive learning rates
        for iteration in range(self.n_iter):
            # Compute Q matrix and gradients
            Q, distances_Y = self._compute_q_matrix_fast(Y)
            grads = self._compute_gradients_fast(P, Q, Y, distances_Y)
            
            # Update with momentum and gains
            gains = (gains + 0.2) * ((grads * (Y - Y_prev)) > 0) + \
                   (gains * 0.8) * ((grads * (Y - Y_prev)) <= 0)
            gains = np.clip(gains, 0.01, 100)
            
            Y_prev = Y.copy()
            Y = Y - self.learning_rate * gains * grads
            Y = Y - np.mean(Y, axis=0)
            
            # Remove early exaggeration after 100 iterations
            if iteration == 100:
                P = P / 4.0
            
            # Print progress less frequently
            if (iteration + 1) % 50 == 0:
                print(f"Iteration {iteration + 1}/{self.n_iter}")
        
        return Y
    
    def _compute_p_matrix_fast(self, D):
        """Fast computation of P matrix using vectorized operations"""
        n_samples = D.shape[0]
        P = np.zeros((n_samples, n_samples))
        beta = np.ones(n_samples)
        target_entropy = np.log(self.perplexity)
        
        for i in range(n_samples):
            # Binary search for beta (precision)
            beta_min = -np.inf
            beta_max = np.inf
            
            for _ in range(20):  # Reduced binary search iterations
                # Compute conditional probabilities
                P_cond = np.exp(-D[i] * beta[i])
                P_cond[i] = 0
                sum_P = np.sum(P_cond)
                
                if sum_P == 0:
                    P_cond = np.zeros_like(P_cond)
                else:
                    P_cond = P_cond / sum_P
                
                # Compute entropy
                entropy = -np.sum(P_cond * np.log(P_cond + 1e-7))
                
                # Update beta based on entropy difference
                entropy_diff = entropy - target_entropy
                if np.abs(entropy_diff) < 1e-5:
                    break
                
                if entropy_diff > 0:
                    beta_min = beta[i]
                    beta[i] = beta[i] * 2 if beta_max == np.inf else (beta[i] + beta_max) / 2
                else:
                    beta_max = beta[i]
                    beta[i] = beta[i] / 2 if beta_min == -np.inf else (beta[i] + beta_min) / 2
            
            P[i] = P_cond
        
        # Symmetrize and normalize P
        P = (P + P.T) / (2 * n_samples)
        P = np.maximum(P, 1e-12)
        
        return P
    
    def _compute_q_matrix_fast(self, Y):
        """Fast computation of Q matrix"""
        sum_Y = np.sum(Y**2, axis=1)
        distances_Y = -2 * np.dot(Y, Y.T) + sum_Y.reshape((-1, 1)) + sum_Y
        inv_distances = 1 / (1 + distances_Y)
        np.fill_diagonal(inv_distances, 0)
        Q = inv_distances / np.sum(inv_distances)
        Q = np.maximum(Q, 1e-12)
        return Q, distances_Y
    
    def _compute_gradients_fast(self, P, Q, Y, distances_Y):
        """Fast gradient computation using vectorized operations"""
        PQ_diff = P - Q
        inv_distances = 1 / (1 + distances_Y)
        grads = 4 * (np.diag(np.sum(PQ_diff * inv_distances, axis=1)).dot(Y) - 
                     (PQ_diff * inv_distances).dot(Y))
        return grads

# Load and preprocess a subset of the MNIST digits dataset
digits = load_digits()
n_samples = 500  # Using only 500 samples for faster execution
indices = np.random.choice(len(digits.data), n_samples, replace=False)
X = digits.data[indices]
y = digits.target[indices]

# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Initialize and run optimized t-SNE
tsne = FastTSNE(n_components=2, perplexity=30.0, learning_rate=200.0, 
                n_iter=300, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

# Visualize results
plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, 
                     cmap='tab10', alpha=0.8)
plt.colorbar(scatter, label='Digit')
plt.title('t-SNE visualization of MNIST digits (500 samples)')
plt.xlabel('t-SNE dimension 1')
plt.ylabel('t-SNE dimension 2')
plt.show()
```
